```python
import matplotlib.pyplot as plt
with open("num.txt")as f:
    data=f.read()
    data=data.split('\n')
    
    x=[row.split(' ')[0] for row in data]
    y=[row.split(' ')[1] for row in data]
    plt.plot(x,y,'o-',color="red")
    
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.title('graph')
    plt.show()
```
